```python
purchase_info = purchase.merge(
    product,
    left_on='product_id',
    right_on='id',
    how='left')
purchase_info['price'] = purchase_info['price'].replace('Free', 0)
purchase_info['price']= purchase_info['price'].astype(float)
purchase_info['cost']=purchase_info['price']*purchase_info['quantity']
purchase_info.head()
```

```python
purchase_info_all = purchase_info.merge(
    product_feedback,
    left_on='product_id',
    right_on='id',
    how='left')
purchase_info_all = purchase_info_all[['user_id','product_id','quantity','product_name','category_name','price','cost','purchase_id','rating','review']]
purchase_info_all.head()
```

```python
customer_info = customer.merge(
    purchase_info_all,
    left_on='id',
    right_on='user_id',
    how='left')
customer_info.head()
```

Задача 1: собрать всю информацию по каждому пользователю в отдельной витрине, а также на основе имеющихся данных посчитать:
+ количество покупок
+ количество потраченных средств
+ средний рейтинг по продукту
+ наивысший продукт по рейтингу
+ самый популярный тип продукта
+ количество отзывов

```python
# количество покупок
total_cost = customer_info.groupby('name')['quantity'].sum().reset_index()
total_cost.columns = ['name', 'quantity']

print(total_cost.head(10))
```

```python
# количество потраченных средств
total_money = customer_info.groupby('name')['cost'].sum().reset_index()
total_money.columns = ['name', 'cost']

print(total_money.head(10))
```

```python
# средний рейтинг по продукту
avg_rating = customer_info.groupby('name')['rating'].mean().reset_index()
avg_rating.columns = ['name', 'avg rating']

print(avg_rating.head(10))
```

```python
#наивысший продукт по рейтингу
max_rating = customer_info.groupby(['name','product_name'])['rating'].max().reset_index()
max_rating.columns = ['name','product_name', 'max rating']
max_rating = max_rating.dropna(subset=['max rating'])
print(max_rating.head(10))
```

```python
#самый популярный тип продукта
popular_product = customer_info.groupby(['name', 'category_name']).size().reset_index(name='Count')
#most_popular = popular_product.loc[popular_product.groupby('name')['Count'].idxmax()]
print(popular_product.head(10))
```

```python
#количество отзывов
review = customer_info.dropna(subset=['review'])
review = customer_info.groupby(['name', 'review']).size().reset_index(name='Count')
review = review.groupby('name')['Count'].sum().reset_index()
print(review.head(10))
```

Задача 2: написать функцию, которая считает долю трат пользователя по разным  категориям от общих трат пользователя. Категории: Movies, Games, Books.

```python
customer_info.head()
```

```python
def calculate_expense_share(df):
    # Группируем по пользователю и категории, считаем сумму трат
    category_expenses = customer_info.groupby(['name', 'product_name'])['cost'].sum().reset_index()
    
    # Считаем общие траты пользователя
    total_expenses = customer_info.groupby('name')['cost'].sum().reset_index()
    total_expenses.rename(columns={'cost': 'TotalCost'}, inplace=True)
    
    # Объединяем данные по категориям с общими тратами
    merged = pd.merge(category_expenses, total_expenses, on='name')
    
    # Считаем долю трат по категориям
    merged['Share'] = merged['cost'] / merged['TotalCost']
    
    return merged[['User', 'Category', 'Share']]

# Используем функцию
expense_share_df = calculate_expense_share(customer_info)
print(expense_share_df)
```
